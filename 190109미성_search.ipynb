```
import pandas as pd
import re
```

```
from google.colab import drive 
drive.mount('/content/gdrive')

search1 = pd.read_csv('gdrive/My Drive/lpoint2.csv', engine='python', encoding='cp949') 
master_brand = pd.read_csv('gdrive/My Drive/master_brand.csv', engine='python', encoding='cp949')
```

```
search1
# Client ID, Session ID를 포함한 검색어의 data table
```

```
# Keyword name에서 unique한 것만 추출

KWD_unique = list(set(search1["KWD_NM"]))
len(KWD_unique)
```

```
original_master_brand = master_brand[:]
original_master_brand
# 브랜드명과 분류명(대, 중, 소)을 포함한 data table
```

```
# brand name 전처리 ("[]", ".", "(주)" 등 제거)

myBRD_NM = []
p1 = re.compile("[(].*[)]")
p2 = re.compile("\W")

for brd_nm in master_brand["PD_BRA_NM"]:
  b1 = p1.sub("", brd_nm)
  b2 = p2.sub("", b1)
  myBRD_NM.append(b2)
  
master_brand["PD_BRA_NM"] = myBRD_NM
master_brand
```

```
# brand name에서 unique한 것만 추출

BRD_unique = list(set(master_brand["PD_BRA_NM"]))
len(BRD_unique)
```

```
# 분류명에서 unique한 것만 추출 (2개 이상이 묶여있는 것 분리)

for clac_nm in set(master_brand["CLAC1_NM"]):
  if "/" in clac_nm: print(clac_nm)
```

```
CLAC1, CLAC2, CLAC3 = [], [], []

for clac1 in set(master_brand["CLAC1_NM"]):
  CLAC1 += clac1.split("/")
  
for clac2 in set(master_brand["CLAC2_NM"]):
  CLAC2 += clac2.split("/")

for clac3 in set(master_brand["CLAC3_NM"]):
  CLAC3 += clac3.split("/")

CLAC1_unique = list(set(CLAC1))
CLAC2_unique = list(set(CLAC2))
CLAC3_unique = list(set(CLAC3))
```

### search1의 Keyword Name에서 Brand Name 분리

```
# 대소문자 통일

for (i, kwd_nm) in enumerate(KWD_unique):
  KWD_unique[i] = kwd_nm.upper()

for (i, brd_nm) in enumerate(BRD_unique):
  BRD_unique[i] = brd_nm.upper()
```

```
# brand name이 분류명과 같은 것 제거

for brd_nm in BRD_unique:
  if brd_nm in (CLAC1_unique + CLAC2_unique + CLAC3_unique):
    BRD_unique.remove(brd_nm)
    print(brd_nm)
```

```
BRD_unique.remove('여성')
BRD_unique.remove('크로스')
```

```
### search1의 Keyword Name에서 Brand Name 분리

myBRD, myKWD = [], []
p = re.compile('(\w*) .*')

for kwd in KWD_unique:
  
  if kwd in BRD_unique: brd = kwd
  elif p.match(kwd):
    g1 = p.match(kwd).group(1)
    brd = g1 if g1 in BRD_unique else ""
  else:
    brd = ""
    for brd_nm in BRD_unique:
      if kwd.find(brd_nm) == 0 and len(brd_nm) > max(1, len(brd)):
        brd = brd_nm
      
  myBRD.append(brd)
  myKWD.append(kwd.replace(brd, "").lstrip())   # brand name 떼고 왼쪽 공백 제거
```

```
mysearch1 = pd.DataFrame({'KWD':list(KWD_unique), 'myBRD':myBRD, 'myKWD':myKWD})
mysearch1
```

## search1을 상품군별로 분류

```
mysearch1_1 = mysearch1[mysearch1["myKWD"] != ""]    # keyword = brand name 인 것 제외
len(mysearch1_1)
```

### 1차 분류: keyword가 CLAC_NM를 포함하는 것 분류

```
CLAC3_unique.remove("유아")
```

```
myCLAC1, myCLAC2, myCLAC3 = [], [], []

for kwd in mysearch1_1["myKWD"]:
  c1, c2, c3 = "", "", ""
  
  if kwd in CLAC2_unique: c2 = kwd
  elif kwd in CLAC3_unique: c3 = kwd
  elif kwd in CLAC1_unique: c1 = kwd

  else:
    for clac2 in CLAC2_unique:
      if clac2 in kwd: c2 = clac2
      
    if c1 == "" and c2 == "" and c3 == "":
      for clac3 in CLAC3_unique:
        if clac3 in kwd and len(clac3) > 1: c3 = clac3
      
      if c3 == "":
        for clac1 in CLAC1_unique:
          if clac1 in kwd: c1 = clac1
            
  myCLAC1.append(c1)
  myCLAC2.append(c2)
  myCLAC3.append(c3)
```

```
mysearch1_2 = pd.DataFrame({'KWD':mysearch1_1["KWD"], 'myBRD':mysearch1_1["myBRD"], 'myKWD':mysearch1_1["myKWD"],
                            'myCLAC1_NM':myCLAC1, 'myCLAC2_NM':myCLAC2, 'myCLAC3_NM':myCLAC3})
mysearch1_2 = mysearch1_2[['KWD', 'myBRD', 'myKWD', 'myCLAC1_NM', 'myCLAC2_NM', 'myCLAC3_NM']]
```

```
mysearch1_2[mysearch1_2.myCLAC2_NM!=""].append(mysearch1_2[mysearch1_2.myCLAC3_NM!=""]).append(mysearch1_2[mysearch1_2.myCLAC1_NM!=""])
```

### 2차 분류: "여성", "남성", "아동" 관련 단어 분리

```
for clac_nm in (CLAC2_unique):
  if "여성" in clac_nm or "남성" in clac_nm or "유아" in clac_nm or "남아" in clac_nm or "여아" in clac_nm:
    print(clac_nm)
# '여성', '남성', '유아', '남아', '여아'로 분류되어있는 CLAC names
```

```
# 1차 분류에서 분류되지 않은 것 중 '여성', '남성', '아동', '여아', '남아' 분리

mysearch1_3 = mysearch1_2[mysearch1_2["myCLAC1_NM"]==""][mysearch1_2["myCLAC2_NM"]==""][mysearch1_2["myCLAC3_NM"]==""][["KWD", "myBRD", "myKWD", 'myCLAC1_NM', 'myCLAC2_NM', 'myCLAC3_NM']]

myGROUP, myKWD = [], []
for kwd in mysearch1_3["myKWD"]:
  if "여성" in kwd or "여자" in kwd:
    myKWD.append(kwd.replace("여성", "").replace("여자", "").strip())
    myGROUP.append("여성")
  elif "남성" in kwd or "남자" in kwd:
    myKWD.append(kwd.replace("남성", "").replace("남자", "").strip())
    myGROUP.append("남성")
  elif "아동" in kwd or "유아" in kwd or "키즈" in kwd:
    myKWD.append(kwd.replace("아동", "").replace("유아", "").replace("키즈", "").strip())
    myGROUP.append("아동")
  elif "남아" in kwd:
    myKWD.append(kwd.replace("남아", "").strip())
    myGROUP.append("남아")
  elif "여아" in kwd:
    myKWD.append(kwd.replace("여아", "").strip())
    myGROUP.append("여아")
  else:
    myKWD.append(kwd)
    myGROUP.append("")
    
mysearch1_3["myGROUP"], mysearch1_3["myKWD"] = myGROUP, myKWD
mysearch1_3 = mysearch1_3[["KWD", "myBRD", "myGROUP", "myKWD", 'myCLAC1_NM', 'myCLAC2_NM', 'myCLAC3_NM']]
```

```
# 분류명에서 '여성', '남성', '아동', '여아', '남아' 분리

CLAC1_unique2, CLAC2_unique2, CLAC3_unique2 = [], [], []

for clac in CLAC1_unique:
  if "여성" in clac or "남성" in clac or "유아동" in clac:
    CLAC1_unique2.append(clac.replace("여성", "").replace("남성", "").replace("유아동", ""))
  else: CLAC1_unique2.append(clac)

for clac in CLAC2_unique:
  if "여성" in clac or "남성" in clac or "유아" in clac or "여아" in clac or "남아" in clac:
    CLAC2_unique2.append(clac.replace("여성", "").replace("남성", "").replace("유아동", "").replace("유아", "").replace("여아", "").replace("남아", ""))
  else: CLAC2_unique2.append(clac)
    
for clac in CLAC3_unique:
  if "여성" in clac or "남성" in clac or "유아" in clac or "아동" in clac or "여아" in clac or "남아" in clac:
    CLAC3_unique2.append(clac.replace("여성", "").replace("남성", "").replace("유아동", "").replace("유아용", "").replace("영유아", "").replace("유아", "").replace("아동용", "").replace("아동", "").replace("여아", "").replace("남아", ""))
  else: CLAC3_unique2.append(clac)

CLAC1_unique2 = list(set(CLAC1_unique2))
CLAC2_unique2 = list(set(CLAC2_unique2))
CLAC3_unique2 = list(set(CLAC3_unique2))
```

```
for (i, clac) in enumerate(CLAC2_unique2):
  if clac == "화": CLAC2_unique2[i] = "신발"
```

```
### 2차 분류

myCLAC1, myCLAC2, myCLAC3 = [], [], []

for kwd in mysearch1_3["myKWD"]:
  c1, c2, c3 = "", "", ""
  
  if kwd in CLAC2_unique2: c2 = kwd
  elif kwd in CLAC3_unique2: c3 = kwd
  elif kwd in CLAC1_unique2: c1 = kwd

  else:
    for clac2 in CLAC2_unique2:
      if clac2 in kwd: c2 = clac2
      
    if c1 == "" and c2 == "" and c3 == "":
      for clac3 in CLAC3_unique2:
        if clac3 in kwd and len(clac3) > 1: c3 = clac3
      
      if c3 == "":
        for clac1 in CLAC1_unique2:
          if clac1 in kwd: c1 = clac1
            
  myCLAC1.append(c1)
  myCLAC2.append(c2)
  myCLAC3.append(c3)
```

```
mysearch1_4 = pd.DataFrame({'KWD':mysearch1_3["KWD"], 'myBRD':mysearch1_3["myBRD"], 'myGROUP':mysearch1_3["myGROUP"],
                            'myKWD':mysearch1_3["myKWD"], 'myCLAC1_NM':myCLAC1, 'myCLAC2_NM':myCLAC2, 'myCLAC3_NM':myCLAC3})
mysearch1_4 = mysearch1_4[['KWD', 'myBRD', 'myGROUP', 'myKWD', 'myCLAC1_NM', 'myCLAC2_NM', 'myCLAC3_NM']]
```

```
mysearch1_4[mysearch1_4.myCLAC2_NM!=""].append(mysearch1_4[mysearch1_4.myCLAC3_NM!=""]).append(mysearch1_4[mysearch1_4.myCLAC1_NM!=""])
```
